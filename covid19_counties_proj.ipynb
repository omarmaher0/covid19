```python
import matplotlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('coviddeathbycountry.csv')
```

We have 3 columns
Country-deaths in country-cases(i think patients)

```python
df.head()
```

```python
df.tail()
```

```python
#in column Deaths you will see sign(_)
#I think it's meaning 0 then i replace it with 0
df["Deaths"] = df['Deaths'].replace("—","0")
```

```python
#Done replaced
df.tail()
```

```python
df.shape
```

```python
#no any null values
df.info()
```

```python
#for numerical data
df.describe()
```

```python
#for categorical data
#column Deaths is categorical because all numbers include "num"(double quote)
df.describe(include=['O'])
```

```python
#I changed the column Deaths to int because it's a numeric data
df['Deaths']=df['Deaths'].astype(int)
```

```python
#for categorical data
df.describe(include=['O'])
```

```python
#no any duplicate values (duplicate means rows have same values)
df.duplicated().sum()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
#you will see more clearly max 5 countries have more deaths in the world
df[['Country','Deaths']].groupby(['Country'],as_index=False).max().sort_values(by='Deaths',ascending=False)
```

```python
#you will see more clearly max 5 countries have more cases(patients) in the world
df[['Country','Cases']].groupby(['Country'],as_index=False).max().sort_values(by='Cases',ascending=False)
```
